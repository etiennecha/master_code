## Price data

Price records collected on a daily basis are aggregagted in a dictionary, the keys of which are fairly explicit:

```python
path_diesel_price = os.path.join(path_dir_built_json, 'master_price_diesel.json')
master_price = dec_json(path_diesel_price)
print master_price.keys()
```

It can then seen that the period studied starts on September 4, 2011 and ends on June 4, 2013, which represents 640 days. However, there are 172 days during which no data were collected (an issue that will be discussed later). The number of gas stations for which there are (at least some) price records is 10,066. 

Station ids are composed by the zip code ('1500' is Ambérieu-en-Bugey) and the registration number ('007' means that it was the seventh station to register within this zip code area... but there are cases of stations creating succesively several accounts, thus "inflating" ids).

```python
print 'First and last dates:', master_price['dates'][0], master_price['dates'][-1]
print 'Length:', len(master_price['dates'])
print 'First and last missing dates:', master_price['missing_dates'][0], master_price['missing_dates'][-1]
print 'Lenth:', len(master_price['missing_dates'])
print 'A few station ids:', master_price['ids'][0:5]
print 'Nb of stations in data:', len(master_price['ids'])
```

'dict_info' is a dictionnary that originally stores all information included in price records but not as dynamic as prices.

- 'brands' (a list of the brand and the day at which it was first observed), 'city' and 'name' are obtained directly from price records.

- 'brand_std', 'code_geo' (insee code), 'code_geo_ardts' (insee_code with specific district codes for big cities) and rank (to be ignored) have been added later on

```python
pprint.pprint(master_price['dict_info']['1500007'])
```

Finally, 'diesel_price' and 'diesel_date' include all prices and the day at which they have been set. The presence of 'diesel_date' allows to fill short missing periods. Indeed if a new price has been set while collection was interupted, it allows to know on which day it was set (originally, it's of course indeed to allow the user of the website to know to which extent price information is reliable..)

```python
pprint.pprint(zip(master_price['diesel_price'][0], master_price['diesel_date'][0])[18:25])
```

Finally, a quicky glance can be cast at a price series:

```python
pylab.rcParams['figure.figsize'] = (15.0, 5.0)
ls_time_index = [pd.to_datetime(date) for date in master_price['dates']]
df_station = pd.DataFrame(master_price['diesel_price'][0],\
                          ls_time_index,\
                          [master_price['ids'][0]])
df_station.plot()
```

The decrease observed end of October 2011 may appear to be suspect given the abnormal size of price changes. However, a quick search on the internet shows that a marketing operation was implemented at this period. On October 28-29, 2011, at a large number of "Intermarche" retailers, gasoline was sold "without margin" ("prix coûtant") according to the brand.

## Distance and competition

Based on gps coordinates available, distances between stations are computed using Haversine formula (check, cf. data processing). Thiis allows to build a list of potential competitors for each gas station.

```python
path_ls_ls_competitors = os.path.join(path_dir_built_json, 'ls_ls_competitors.json')
path_ls_tuple_competitors = os.path.join(path_dir_built_json, 'ls_tuple_competitors.json')
ls_ls_competitors = dec_json(path_ls_ls_competitors)
ls_tuple_competitors = dec_json(path_ls_tuple_competitors)
```

For each gas stations of which the location is known, ids and distance of gas stations located within 10 km are stored in ls_ls_competitors.

## Cost of wholesale gasoline

```python
from BeautifulSoup import BeautifulSoup
path_dir_rotterdam = os.path.join(path_data, 'data_gasoline', 'data_source', 'data_rotterdam')
path_dir_reuters = os.path.join(path_dir_rotterdam, 'data_reuters')
path_xls_reuters_diesel = os.path.join(path_dir_reuters, 'diesel_data_to_import.xls')
path_xml_ecb = os.path.join(path_dir_rotterdam, 'usd.xml')

reuters_diesel_excel_file = pd.ExcelFile(path_xls_reuters_diesel)
df_reuters_diesel = reuters_diesel_excel_file.parse('Feuil1', skiprows = 0, header = 1, parse_dates = True)
df_reuters_diesel.set_index('Date', inplace = True)

ecb_xml_file = open(path_xml_ecb, 'r').read()
soup = BeautifulSoup(ecb_xml_file)
ls_ecb = [[pd.to_datetime(dict(obs.attrs)[u'time_period']), float(dict(obs.attrs)[u'obs_value'])]\
            for obs in soup.findAll('obs')]
df_ecb = pd.DataFrame(ls_ecb, columns = ['Date', 'ECB Rate ED'])
df_ecb.set_index('Date', inplace = True)

index = pd.date_range(start = pd.to_datetime('20110904'),
                      end   = pd.to_datetime('20130604'), 
                      freq='D')
df_cost = pd.DataFrame(None, index = index)
for df_temp in [df_ecb, df_reuters_diesel]:
  for column in df_temp:
    df_cost[column] = df_temp[column]

litre_per_us_gallon = 3.785411784
litre_per_barrel = 158.987295
litre_per_metric_tonne = 1183.5

df_cost['ULSD 10 FOB MED EL'] = df_cost['ULSD 10 FOB MED DT'] / litre_per_metric_tonne / df_cost['ECB Rate ED']
df_cost['ULSD 10 CIF NWE EL'] = df_cost['ULSD 10 CIF NWE DT'] / litre_per_metric_tonne / df_cost['ECB Rate ED']
df_cost['GASOIL 0.2 FOB ARA EL'] = df_cost['GASOIL 0.2 FOB ARA DT'] / litre_per_metric_tonne / df_cost['ECB Rate ED']
```
