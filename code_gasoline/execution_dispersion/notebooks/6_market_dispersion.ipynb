## Introduction



In the seminal paper of the consumer search and price dispersion literature, Stigler (1961) cited the example of a specific car model sold by various retailers within a suburb at different prices. This was meant to support the idea that the existence of price dispersion reflects the "ignorance in the market", namely consumers' lack of information about prices. On the one hand, there are of course many alternative explanations that one may want to consider to account for such dispersion, such as heterogeneity in the perceived reliability of retailers or the fact that cars typically end up being sold with rebates. Yet, this example could be seen as just another example of the "law of one price" appearing to be irrelevant and too loose to describe actual competition.

Even more striking, the literature would later come up with the concept of "rank reversals" i.e. the fact that the ranks of competiting stores in the price distribution may vary over time. Stores would then appear to be eratically cheap or expensive compared to competitors, leaving less room for explanations in terms of differentiation.

Chandra and Tappata (2011) seek to test the relevance of the model to account for retail gasoline prices, a market which arguably presents a good fit with standard theoretical models of price dispersion. Seeking to overcome this issues leads to consider various statistics accounting for price dispersion as well as "leadership" i.e. pattern of price adjustements and thus draw contours of competition categories.

## Methodology

### Overview

A significant number of statistics can be computed to compare prices at two stations:

- average spread : a priori around 0 if no differentiation
- standard deviation of spread : higher in the case of symmetric mixed price equilibra
- average absolute spread: can be different from 0 while average spread is c. 0, in particular in the case of symmetric mixed price equilibra
- frequency of equal prices
- frequency of prices being set by one seller
- frequency of prices being set simultaneously by both sellers (given daily observations
- rank reversal: frequency of days during which the generally cheapest station turns out to be more expensive (and vice versa)
- number of rank reversals over the period and length of rank reversals

In the case of differentiation (current):

- rank reversal: one price series can be normalized (using the average spread) so that statistics can be computed
- leadership: number of days with simultaneous price changes / price changes followed (regardless of variation value and sign)
- stability: s(TODO) frequency of the most common spread between prices?

Difficulties:

- several potentially relevant competitors
- no absolute value threshold...

A proxy for leadership can be obtained by counting the number of times a seller sets a price that is then matched by the other seller. A more sophisticated version can allow for such a pattern with different price levels. Different outcomes of competition can then be characterized:

- Tough competition/candid collusion
  - average spread and average absolute spread close to 0
  - price changes essentially occur simultaneously (as far as we can conclude from daily observations)

- Tough competition or imperfect/complex collusion?
  - average spread and average absolute spread close to 0
  - price changes not simultaneous but no clear leadership

- Leadership (roles can change over time): competition or collusion?
  - average spread and average absolute spread not 0
  - a seller essentially matches prices set by the other seller 

- "Deterministic" dispersion
  - average spread close to 0, average absolute spread not 0
  - different price levels, predictible alternation of each seller being cheaper or more expensive

- "Random" dispersion
  - average spread close to 0, average absolute spread not 0
  - different price levels, chaotic alternation of each sellers being cheaper or more expensive

Each scenario can then be examined with a difference in price levels (normalization of prices of one seller).

### Implementation

All pairs of gas stations separated by less than 10 km are first considered and split on an average spread criterion (ARGUE... data!):
    
- pairs which exhibit an average spread below 0.02€/l are kept as non significantly differentiated and likely close competitors.
- pairs which exhibit an average spread above 0.02€/l are considered as signficantly differentiated and potentially not close competitors.


```python
ls_ls_leadership = []
for indiv_ind, indiv_id in enumerate(master_price['ids']):
  ar_price_1 = np.array(df_price[indiv_id])
  ls_leadership = []
  for comp_id, distance in ls_ls_competitors[indiv_ind][0:10]:
    ar_price_2 = np.array(df_price[comp_id])
    len_spread, len_same, ls_1_lead, ls_2_lead, ls_ctd_1, ls_ctd_2 = get_two_firm_similar_prices(ar_price_1, ar_price_2)
    if len(ls_1_lead) != len(ls_ctd_1):
      print 'i, pbm 1',
      break
    if len(ls_2_lead) != len(ls_ctd_2):
      print 'i, pbm 2'
      break
    ls_leadership.append((len_spread, len_same, ls_1_lead, ls_2_lead, ls_ctd_1, ls_ctd_2))
  ls_ls_leadership.append(ls_leadership)

#df_price[['1500006', '1500003']].ix[575:585]
```

```python
ls_ls_leadership_light = []
for ls_leadership in ls_ls_leadership:
  ls_leadership_light = []
  for (len_spread, len_same, ls_1_lead, ls_2_lead, ls_ctd_1, ls_ctd_2) in ls_leadership:
    ls_leadership_light.append((len_spread, len_same, len(ls_1_lead), len(ls_2_lead)))
  ls_ls_leadership_light.append(ls_leadership_light)

ls_station_leadership_results = []
for ls_leadership_light in ls_ls_leadership_light:
  station_leadership_results = []
  ls_leadership_light_alt = zip(*ls_leadership_light)
  for ls_light in ls_leadership_light_alt:
    if ls_light:
      station_leadership_results.append(max(ls_light))
    else:
      station_leadership_results.append(None) 
  ls_station_leadership_results.append(station_leadership_results)
        
print len(ls_ls_leadership)
print master_price['ids'].index('1500006')

import pprint
pprint.pprint(ls_ls_leadership_light[1])

ls_columns = ['len_spread', 'len_same', 'len_before', 'len_after']
df_leadership_results = pd.DataFrame(ls_station_leadership_results, index=master_price['ids'], columns = ls_columns)
print df_leadership_results.ix[0:10]
#df_price[['1200003', '1200001']][300:500].plot()
```
