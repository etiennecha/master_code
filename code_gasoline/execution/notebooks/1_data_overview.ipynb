## Price data

Price records collected on a daily basis are aggregagted in a dictionary:

```python
path_diesel_price = os.path.join(path_dir_built_json, 'master_price_diesel.json')
master_price = dec_json(path_diesel_price)
print master_price.keys()
```

The period studied starts on September 4, 2011 and ends on June 4, 2013, which represents 640 days. There are yet 172 days during which no data were collected (an issue that will be discussed later). 

The number of gas stations for which there are (at least some) price records is 10,066. 

Station ids are composed by the zip code ('1500' is Ambérieu-en-Bugey) and the registration number ('007' means that it was the seventh station to register within this zip code area... but there are cases of stations creating succesively several accounts, thus "inflating" ids).

```python
print 'First and last dates:', master_price['dates'][0], master_price['dates'][-1]
print 'Length:', len(master_price['dates'])
print 'First and last missing dates:', master_price['missing_dates'][0], master_price['missing_dates'][-1]
print 'Lenth:', len(master_price['missing_dates'])
print 'A few station ids:', master_price['ids'][0:5]
print 'Nb of stations in data:', len(master_price['ids'])
```

'dict_info' is a dictionnary that originally stores all information collected on daily basis but not as changing as price:.

- 'brands' (a list of the brand and the day at which it was first observed), 'city' and 'name' are obtained directly from daily records

- 'brand_std', 'code_geo' (insee code), 'code_geo_ardts' (insee_code with specific district codes for big cities) and rank (to be ignored) have been added later on

```python
pprint.pprint(master_price['dict_info']['1500007'])
```

Finally, 'diesel_price' and 'diesel_date' include respectively all station prices and the day at which they have been set. The presence of 'diesel_date' allows to fill short missing periods: if a new price is observed after some interruption, the date of the price change being known, the series can completed.

```python
pprint.pprint(zip(master_price['diesel_price'][0], master_price['diesel_date'][0])[18:25])
```

Finally, a quicky glance can be cast at a price series:

```python
pylab.rcParams['figure.figsize'] = (15.0, 5.0)
ls_time_index = [pd.to_datetime(date) for date in master_price['dates']]
df_price = pd.DataFrame(np.array(master_price['diesel_price'], np.float64).T,
                         index = ls_time_index,
                         columns = master_price['ids'])
df_price['1500007'].plot()
```

The decrease observed end of October 2011 may appear to be suspect given the abnormal size of price changes. However, a quick search on the internet shows that a marketing operation was implemented at this period. On October 28-29, 2011, at a large number of "Intermarche" retailers, gasoline was sold "without margin" ("prix coûtant") according to the brand.

## Distance and competition

Based on gps coordinates available, distances between stations are computed using Haversine formula (check, cf. data processing). Thiis allows to build a list of potential competitors for each gas station.

```python
path_ls_ls_competitors = os.path.join(path_dir_built_json, 'ls_ls_competitors.json')
path_ls_tuple_competitors = os.path.join(path_dir_built_json, 'ls_tuple_competitors.json')
ls_ls_competitors = dec_json(path_ls_ls_competitors)
ls_tuple_competitors = dec_json(path_ls_tuple_competitors)
```

For each gas stations of which the location is known, ids and distance of gas stations located within 10 km are stored in ls_ls_competitors.

```python
print ls_ls_competitors[0]
```

The price series of a gas station can then be displayed together with the price series at any competitors within a certain number of km.

```python
ls_ls_market_ids = [master_price['ids'][0]] + [comp[0] for comp in ls_ls_competitors[0] if comp[1] < 5]
df_price[ls_ls_market_ids].plot()
```

Additionally, the ids of each pair of gas stations separated by less than 10 km are stored in ls_tuple_competitors. With the 10,066 gas stations present in the data (TODO: how many location unknown?), 97,030 pairs are formed with less than 10 km distance.

## Cost of wholesale gasoline

```python
from BeautifulSoup import BeautifulSoup
path_dir_rotterdam = os.path.join(path_data, 'data_gasoline', 'data_source', 'data_rotterdam')
path_dir_reuters = os.path.join(path_dir_rotterdam, 'data_reuters')
path_xls_reuters_diesel = os.path.join(path_dir_reuters, 'diesel_data_to_import.xls')
path_xml_ecb = os.path.join(path_dir_rotterdam, 'usd.xml')

reuters_diesel_excel_file = pd.ExcelFile(path_xls_reuters_diesel)
df_reuters_diesel = reuters_diesel_excel_file.parse('Feuil1', skiprows = 0, header = 1, parse_dates = True)
df_reuters_diesel.set_index('Date', inplace = True)

ecb_xml_file = open(path_xml_ecb, 'r').read()
soup = BeautifulSoup(ecb_xml_file)
ls_ecb = [[pd.to_datetime(dict(obs.attrs)[u'time_period']), float(dict(obs.attrs)[u'obs_value'])]\
            for obs in soup.findAll('obs')]
df_ecb = pd.DataFrame(ls_ecb, columns = ['Date', 'ECB Rate ED'])
df_ecb.set_index('Date', inplace = True)

index = pd.date_range(start = pd.to_datetime('20110904'),
                      end   = pd.to_datetime('20130604'), 
                      freq='D')
df_cost = pd.DataFrame(None, index = index)
for df_temp in [df_ecb, df_reuters_diesel]:
  for column in df_temp:
    df_cost[column] = df_temp[column]

litre_per_us_gallon = 3.785411784
litre_per_barrel = 158.987295
litre_per_metric_tonne = 1183.5

df_cost['ULSD 10 FOB MED EL'] = df_cost['ULSD 10 FOB MED DT'] / litre_per_metric_tonne / df_cost['ECB Rate ED']
df_cost['ULSD 10 CIF NWE EL'] = df_cost['ULSD 10 CIF NWE DT'] / litre_per_metric_tonne / df_cost['ECB Rate ED']
df_cost['GASOIL 0.2 FOB ARA EL'] = df_cost['GASOIL 0.2 FOB ARA DT'] / litre_per_metric_tonne / df_cost['ECB Rate ED']
```

Neglecting volumes and variations in taxes accross regions, the average price before tax can be plotted together with the two best wholesale diesel cost proxies.

End of August 2013, a temporary cut in gasoline prices (vertical red line) was decided by the government of the recently elected president Francois Hollande. The target announced was a decrease of 6c€/l achieved half through tax reduction (bearing on TICPE, which is included in the computation of VAT) and half through gas station margin reduction. 

Tax was increased again by 1c€/l on November 30, and by 0.5c€/l twice on December 11 and 21. It was finally set back at its original level on January 11, 2013.

```python
df_cost['avg_diesel_price_ttc'] = df_price.mean(axis = 1)
df_cost['avg_diesel_price_ht'] = df_price.mean(axis = 1)

df_cost['avg_diesel_price_ht'].ix[:'2012-08-30'] =\
  df_cost['avg_diesel_price_ht'].ix[:'2012-08-30'] / 1.196 - 0.4419

df_cost['avg_diesel_price_ht'].ix['2012-08-31':'2012-11-30'] =\
  df_cost['avg_diesel_price_ht'].ix['2012-08-31':'2012-11-30'] / 1.196 - 0.4419 + 0.03

df_cost['avg_diesel_price_ht'].ix['2012-12-01':'2012-12-11'] =\
  df_cost['avg_diesel_price_ht'].ix['2012-12-11':'2012-12-11'] / 1.196 - 0.4419 + 0.02
    
df_cost['avg_diesel_price_ht'].ix['2012-12-11':'2012-12-21'] =\
  df_cost['avg_diesel_price_ht'].ix['2012-12-11':'2012-12-21'] / 1.196 - 0.4419 + 0.015

df_cost['avg_diesel_price_ht'].ix['2012-12-21':'2013-01-11'] =\
  df_cost['avg_diesel_price_ht'].ix['2012-12-21':'2013-01-11'] / 1.196 - 0.4419 + 0.01

df_cost['avg_diesel_price_ht']['2013-01-11':] =\
  df_cost['avg_diesel_price_ht'].ix['2013-01-11':] / 1.196 - 0.4419

ax = df_cost[['avg_diesel_price_ht', 'ULSD 10 FOB MED EL', 'ULSD 10 CIF NWE EL']].plot()
ax.axvline(x=pd.to_datetime('2012-08-31'), linewidth=1, color='r')
ax.axvline(x=pd.to_datetime('2013-01-11'), linewidth=1, color='r')
```

A gross margin can then be computed and plotted. 

```python
df_cost['avg_diesel_gross_margin'] = df_cost['avg_diesel_price_ht'] - df_cost['ULSD 10 CIF NWE EL']
ax = df_cost['avg_diesel_gross_margin'].plot()
ax.axvline(x=pd.to_datetime('2012-08-31'), linewidth=1, color='r')
ax.axvline(x=pd.to_datetime('2013-01-11'), linewidth=1, color='r')
df_cost.index[0].to_pydatetime()
```


The decrease of the gross margin as early as july i.e. before the measure taken by the government appears to be consistant with observations of the governmental body in charge of monitoring margins:

http://www.economie.gouv.fr/files/files/directions_services/dgccrf/concurrence/prix/mensuel_2012/OPm_petrol_juillet12.pdf

```python
avg_ttc_june = df_cost['avg_diesel_price_ttc'].ix['2012-06-01':'2012-06-30'].mean()
avg_ttc_july = df_cost['avg_diesel_price_ttc'].ix['2012-07-01':'2012-07-31'].mean()
print 'Avg june, july, diff', avg_ttc_june, avg_ttc_july, avg_ttc_july-avg_ttc_june
print 'Avg june, july, diff OPM', 1.342, 1.378, 1.378-1.342
gm_june =df_cost['avg_diesel_gross_margin'].ix['2012-06-01':'2012-06-30'].mean()
gm_july = df_cost['avg_diesel_gross_margin'].ix['2012-07-01':'2012-07-31'].mean()
print gm_july - gm_june
print (gm_july-gm_june)/gm_june # -22% according to OPM (similar variation but smaller gross margin in value)

ax = df_cost['avg_diesel_gross_margin'].ix['2012-04-30':'2013-04-30'].plot()
ax.axvline(x=pd.to_datetime('2012-08-31'), linewidth=1, color='r')
ax.axvline(x=pd.to_datetime('2013-01-11'), linewidth=1, color='r')
```
